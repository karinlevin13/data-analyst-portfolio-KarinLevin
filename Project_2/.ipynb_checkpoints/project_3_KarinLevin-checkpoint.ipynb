# Step 2. Make reports and calculate metrics <a name="step2"></a>

### Product <a name="subparagraph1"></a>

#### How many people use it every day, week, and month?

```python
users_data=visits
users_data['session_year']  = users_data['start_ts'].dt.year
users_data['session_month'] = users_data['start_ts'].dt.month
users_data['session_week']  = users_data['start_ts'].dt.week
users_data['session_date'] = users_data['start_ts'].dt.date
```

    DAU — the number of daily active (unique) users
    WAU — the number of weekly active users
    MAU — the number of monthly active users

```python
dau_total = users_data.groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total = users_data.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'}).mean()
mau_total = users_data.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'}).mean()
print('the number of daily active (unique) users are ',int(dau_total))
print('the number of weekly active users are ',int(wau_total)) 
print('the number of monthly active users are',int(mau_total))
```

## How many sessions are there per day? (One user might have more than one session.)

```python
sessions_per_user = users_data.groupby(['session_year', 'session_date']).agg(
    {'uid': ['count', 'nunique']}
)
sessions_per_user.columns = ['n_sessions', 'n_users']
sessions_per_user['sessions_per_user'] = (
    sessions_per_user['n_sessions'] / sessions_per_user['n_users']
)
print(sessions_per_user) 
```

```python
print(sessions_per_user['sessions_per_user'].median())
```

###  Conclusion

This table, as well as the median, shows that all users generally visit the site only once a day.

### What is the length of each session?

```python
users_data['session_duration_sec'] = (users_data['end_ts'] - users_data['start_ts']).dt.seconds
logs_with_duration = users_data.loc[(users_data['session_duration_sec']>0)]# Make a data slice
print(logs_with_duration['session_duration_sec'].mean()) 
```

```python


#plotting
plt.title("the distribution")
logs_with_duration['session_duration_sec'].hist(bins=60) 
plt.show()

print('ASL is ',logs_with_duration['session_duration_sec'].mode()) 
#Average session length, or ASL, is the amount of time users spend with a product 
#in the average session.
```

###  Conclusion

These results show us that either users come in, buy nothing and leave the site within the first 60 seconds, or when a user buys something, he spends much more time, about 10 minutes or more.

### How often do users come back?

```python
user_activity=pd.DataFrame({'user_id':users_data['uid'],'activity_date':users_data['start_ts']})

user_activity['activity_date'] = pd.to_datetime(user_activity['activity_date'])
first_activity_date = user_activity.groupby(['user_id'])['activity_date'].min()
first_activity_date.name = 'first_activity_date'
user_activity = user_activity.join(first_activity_date, on='user_id')

user_activity['activity_month'] = pd.to_datetime(
    user_activity['activity_date'], unit='d'
) - pd.to_timedelta(user_activity['activity_date'].dt.month, unit='d')
user_activity['first_activity_month'] = pd.to_datetime(
    user_activity['first_activity_date'], unit='d'
) - pd.to_timedelta(
    user_activity['first_activity_date'].dt.month, unit='d'
) 
user_activity['cohort_lifetime'] = (
    user_activity['activity_month'] - user_activity['first_activity_month']
)
user_activity['cohort_lifetime'] = user_activity[
    'cohort_lifetime'
] / np.timedelta64(1, 'M')
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'].astype(int)

cohorts =(
    user_activity.groupby(['first_activity_month', 'cohort_lifetime'])
    .agg({'user_id': 'nunique'})
    .reset_index()
) 
initial_users_count =cohorts[cohorts['cohort_lifetime'] == 0][
    ['first_activity_month', 'user_id']
] # Build the data frame here
initial_users_count =  initial_users_count.rename(
    columns={'user_id': 'cohort_users'}
) # Rename the data frame column
cohorts =  cohorts.merge(initial_users_count, on='first_activity_month') # Join the data frames cohorts and initital_users_count
cohorts['retention'] = cohorts['user_id'] / cohorts['cohort_users']
# Calculate retention rate
cohorts['first_activity_month_']=cohorts['first_activity_month'].dt.to_period("m")
retention_pivot = cohorts.pivot_table(
    index='first_activity_month_',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
).fillna(0).round(2)# Compile pivot table


```

```python

plt.figure(figsize=(15, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, cmap="YlGnBu")
plt.show()
```

###  Conclusion

The data shows us that users make purchases mostly immediately after registering and then hardly visit the site.

```python
#To build a cohort report, we need to determine the date and time of the first session of each user. 
first_visits = visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_session_start_ts']

#adding the columns of the date and month for convenience of grouping
first_visits['first_session_dt'] = first_visits['first_session_start_ts'].dt.date
first_visits['first_session_month'] = first_visits['first_session_start_ts'].astype('datetime64[M]')

#make columns session_month normal
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')

#adding information about the first visit to the dataframe with visitors 
vis = pd.merge(first_visits, visits, on = 'uid')

#counting how much time has passed between each session and the first user session 
vis['age_months'] = ((vis['session_month'] - vis['first_session_month']) / np.timedelta64(1,'M')).round().astype('int')

#Cohort analysis by the number of visitors. 
cohorts = vis.pivot_table(index='first_session_month',
                  columns='age_months',
                  values='uid',
                  aggfunc='nunique')
cohorts.fillna('')


#Retention Rate
retention = pd.DataFrame()
for col in cohorts.columns:
    retention = pd.concat([retention, cohorts[col]/cohorts[0]], axis=1)
retention.columns = cohorts.columns
retention.index = [str(x)[0:10] for x in retention.index]
plt.figure(figsize=(13, 9))
sns.heatmap(retention, annot=True, fmt='.1%', linewidths=1, linecolor='grey',  vmax=0.1, cbar_kws= {'orientation': 'horizontal'} 
            ).set(title = 'Retention Rate')
plt.show()
```

## Sales <a name="subparagraph2"></a>

#### When do people start buying? 

```python
costs['dt'].describe() 
```

```python
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]') 
costs['month_'] =pd.to_datetime(costs['dt']).dt.strftime('%Y-%m') 
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
first_orders.head() 
```

```python
cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head() 
```

```python
orders_ = pd.merge(orders,first_orders, on='uid')
orders_.head() 
```

```python
cohorts = orders_.groupby(['first_order_month','order_month']).agg({'revenue': ['sum','median','count']}).reset_index()
cohorts.head() 
```

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report.columns=['first_order_month','n_buyers','order_month','revenue_sum','revenue_median','revenue_count']
report.head() 
```

```python
margin_rate = 0.5

report['gp'] = report['revenue_sum'] * margin_rate
report['age'] = (
    report['order_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')

display(report)
```

```python


first=orders.merge(visits,on='uid')
first=first.groupby('uid')[['buy_ts','start_ts']].min()
first['delta']=first['buy_ts']-first['start_ts']
first.columns=['first_order','first_visit','delta']
display(first)

print(first['delta'].median())
first.describe()
```

###  Conclusion

This table shows us that the profit from all cohorts in the first months is approximately the same, and then the profit falls after first month in each cohort by about 10 times.because almost everyone did their orders no more that 16 minutes after first visit.We can see it also on count of orders per months.

#### How much money do they bring? (LTV)

```python
report['ltv'] = report['gp'] / report['n_buyers']
report["first_order_month_"]=pd.to_datetime(report["first_order_month"]).dt.strftime('%Y-%m')
output = report.pivot_table(
    index='first_order_month_', columns='age', values='ltv', aggfunc='mean'
).round()

output.round(0).fillna('') 


plt.figure(figsize=(20, 8))
plt.title('LTV')
sns.heatmap(output, annot=True, fmt=".2f", linewidths=.5)
plt.xlabel('Age of cohorts')
plt.ylabel('Cohorts')
plt.show()
```

###  Conclusion

LTV shows us that the profit from all cohorts in the first month is 2.5-3 dollars, and then 0, that is, buyers place all orders only in the first month after registration, and then do not go to the site

```python
#adding the day and month of the order to calculate the total revenue for these periods 
orders['buy_dt'] = orders['buy_ts'].dt.date
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

#Finding the time of the first purchase of each customer 
first_orders = orders.groupby('uid').agg({'buy_ts': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_ts']
first_orders['first_order_dt'] = first_orders['first_order_ts'].dt.date
first_orders['first_order_month'] = first_orders['first_order_ts'].astype('datetime64[M]')

#combining data on first purchases with data on first sessions 
buyers = pd.merge(first_visits, first_orders, on='uid')
```

```python
#To answer the question, let's build a cohort report from the date of the first purchase. 
#Find the number of customers in each cohort 
cohort_sizes = buyers.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.rename(columns={'uid': 'n_buyers'}, inplace=True)

#Add the month of the first purchase to the purchase information 
cohorts = pd.merge(orders, buyers, how='inner', on='uid')\
            .groupby(['first_order_month', 'order_month'])\
            .agg({'revenue': 'sum'}).reset_index()

#calculating the age of each cohort 
cohorts['age_month'] = ((cohorts['order_month'] - cohorts['first_order_month']) / np.timedelta64(1,'M')).round()
cohorts.columns = ['first_order_month', 'order_month', 'revenue', 'age_month']

#Add the number of buyers in each cohort to the cohort report  
#and calculate the revenue for each customer. 
cohorts_report = pd.merge(cohort_sizes, cohorts, on = 'first_order_month')
cohorts_report['rev_per_buyer'] = cohorts_report['revenue'] / cohorts_report['n_buyers']

#Age cohort report showing cumulative revenue per customer 

cohorts_age = cohorts_report.pivot_table(
    index='first_order_month', 
    columns='age_month', 
    values='rev_per_buyer', 
    aggfunc='sum'
).cumsum(axis=1)

cohorts_age.round(2).fillna('')

plt.figure(figsize=(13, 9))
sns.heatmap(cohorts_age, annot=True, fmt='.2f', linewidths=1, linecolor='grey', cbar_kws= {'orientation': 'horizontal'} 
            ).set(title ='LTV (by the time of the first purchase )', yticklabels = cohorts_age.index.astype('str'))
plt.show()
```

#### How worthwhile where the investments? (ROI)

per source

```python
#make new table for first visits
first_visits = visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_visits_datetime']
first_visits['first_visits_date'] = first_visits['first_visits_datetime'].dt.date
first_visits['first_visits_month'] = first_visits['first_visits_datetime'].astype('datetime64[M]')

#make new table for first orders
first_orders = orders.groupby('uid').agg({'buy_ts': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_datetime']
first_orders = pd.merge(first_orders, first_visits, on='uid')

#merge tables and add columns
orders_new = pd.merge(orders, first_orders, on='uid')
orders_new['order_month'] = orders_new['buy_ts'].astype('datetime64[M]')
orders_new['first_order_month'] = orders_new['first_order_datetime'].astype('datetime64[M]')

#find first source_id for client
users = visits.sort_values('start_ts').groupby('uid').first()
users = users[['source_id']]

#merge tables
orders_new = pd.merge(orders_new, users, left_on='uid', right_index=True)

#group for uid
cohort_sizes_t = (orders_new.groupby(['source_id', 'first_order_month'])
                            .agg({'uid': 'nunique'})
                            .reset_index())
cohort_sizes_t.columns = ['source_id', 'first_order_month', 'n_buyers']

#group for revenue
cohorts_revenue_t = (orders_new.groupby(['source_id', 'first_order_month', 'order_month'])
                               .agg({'revenue': 'sum'})
                               .reset_index())
cohorts_revenue_t.columns = ['source_id', 'first_order_month', 'order_month', 'gp']

#merge tables and add columns
report_romi = pd.merge(cohort_sizes_t, cohorts_revenue_t, on=['source_id', 'first_order_month'])
report_romi['age'] = (report_romi['order_month'] - report_romi['first_order_month']) / np.timedelta64(1, 'M')
report_romi['age'] = report_romi['age'].round().astype('int')
report_romi['ltv'] = report_romi['gp'] / report_romi['n_buyers']

#add column for costs
costs['costs_month'] = costs['dt'].astype('datetime64[M]')

#add costs and make calculations
report_romi = pd.merge(report_romi,costs.groupby(['source_id', 'costs_month']).agg({'costs': 'sum'}).reset_index(),
                       left_on=['source_id', 'first_order_month'], right_on=['source_id', 'costs_month'])
report_romi = report_romi.drop(['costs_month'], axis = 1) 
report_romi['cac'] = report_romi['costs'] / report_romi['n_buyers']
report_romi['romi'] = report_romi['ltv'] / report_romi['cac']
display(report_romi)
```

```python
report_['romi'] = report_['ltv'] / report_['cac']
output = report_.pivot_table(
    index='first_order_month', columns='age', values='romi', aggfunc='mean'
)
 
output.cumsum(axis=1).round(2).fillna('')  

plt.figure(figsize=(20, 8))
plt.title('ROMI')
sns.heatmap(output, annot=True, fmt=".2f", linewidths=.5,
           yticklabels=output.cumsum(axis=1).round(2).index.strftime('%Y-%m'))
plt.xlabel('Age')
plt.ylabel('Cohorts')
plt.show()
```

ROMI (Marketing Return On Investment) = (ad revenue - ad spend) / ad spend × 100%.


Based on this table, in the months when there were the most orders, that is, in 1, the ROMI from advertising is very low in all cohorts.
